```python
pdf = pd.read_json(datasetPath, lines=True, encoding='utf-8')
pdf.head()
```

```python
# Huggingface Datasets 객체로 변환
hf_dataset = Dataset.from_pandas(pdf)

# 허깅페이스 허브에 업로드
print(f"허깅페이스 허브({hf_repo})에 업로드 시도...")
hf_dataset.push_to_hub(hf_repo, token=hf_token)
print("허깅페이스 업로드 완료!")
```
